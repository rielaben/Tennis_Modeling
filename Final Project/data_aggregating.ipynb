```python
# convert European odds into probabilities

tennis_df['B365W_prob'] = (1/tennis_df['B365W']).round(3)
tennis_df['B365L_prob'] = (1/tennis_df['B365L']).round(3)
```

```python
tennis_df_historical = tennis_df.copy()
```

```python
# set up return DF with our new formatted data
final_columns = ['ATP', 'Location', 'Tournament', 'Date', 
                 'Series', 'Court', 'Surface', 'Round',
                'HR_name', 'HR_rank', 'HR_prob', 
                 'LR_name', 'LR_rank', 'LR_prob']
final_df = pd.DataFrame(columns = final_columns)


# iterate through rows of df, for each row we need to do manipulation
for index, row in tennis_df[5000:5005].iterrows():
    
#     create dictionary to return the new row
    returned_row = {}
    returned_row['ATP'] = row['ATP']
    returned_row['Location'] = row['Location']
    returned_row['Tournament'] = row['Tournament']
    returned_row['Date'] = row['Date']
    returned_row['Year'] = row['Year']
    returned_row['Series'] = row['Series']
    returned_row['Court'] = row['Court']
    returned_row['Surface'] = row['Surface']
    returned_row['Round'] = row['Round']
    
    
#     display original df row
    display(row.to_frame().T)
    
#     get date of the match
    date_of_match = row['Date']
    
#     set up information about the winner and loser
    winner = {}
    winner['name'] = row['Winner']
    winner['rank'] = row['WRank']
    winner['prob'] = row['B365W_prob']
    
    loser = {}
    loser['name'] = row['Loser']
    loser['rank'] = row['LRank']
    loser['prob'] = row['B365L_prob']
    
#     figure out the higher and lower rank
    higher_rank = {}
    lower_rank = {}
    if winner['rank'] < loser['rank']:
        higher_rank = winner
        lower_rank = loser
    else:
        higher_rank = loser
        lower_rank = winner
        
#     add name and rank info to returned df
    returned_row['HR_name'] = higher_rank['name']
    returned_row['HR_rank'] = higher_rank['rank']
    returned_row['HR_prob'] = higher_rank['prob']
    returned_row['LR_name'] = lower_rank['name']
    returned_row['LR_rank'] = lower_rank['rank']
    returned_row['LR_prob'] = lower_rank['prob']
    

# isolate the data before the match played where the higher rank player played (wins and losses)
    higher_rank_name = higher_rank['name']
    print(higher_rank_name)
    print(date_of_match)
    higher_rank_history = tennis_df_historical.loc[(tennis_df_historical['Winner'] == higher_rank_name) | (tennis_df_historical['Loser'] == higher_rank_name)]
    higher_rank_history = higher_rank_history[higher_rank_history['Date'] < date_of_match]
    
#     isolate last 10 games for player
    higher_rank_last_10 = higher_rank_history[-10:]
    
    display(higher_rank_last_10)
    
    final_df = final_df.append(returned_row, ignore_index = True)
```
