```python
import pandas as pd
import numpy as np

pd.set_option('max.colu mns', 100)
```

```python
matches = pd.read_csv('complete_data.csv').drop('Unnamed: 0', axis=1)
matches.head()
```

```python
matches.index = range(len(matches))
matches
```

```python
matches['higher_ranked_odds'] = 1/matches['higher_ranked_prob']
matches['lower_ranked_odds'] = 1/matches['lower_ranked_prob']
```

```python
matches
```

```python
matches['higher_ranked_wpct_on_surface_adj'] = matches['higher_ranked_surface_type_win_pct'] - matches['higher_ranked_all_time_win_pct']
matches['lower_ranked_wpct_on_surface_adj'] = matches['lower_ranked_surface_type_win_pct'] - matches['lower_ranked_all_time_win_pct']
```

```python
matches['higher_ranked_wpct_on_surface_adj'] = matches['higher_ranked_wpct_on_surface_adj'].fillna(0)
matches['lower_ranked_wpct_on_surface_adj'] = matches['lower_ranked_wpct_on_surface_adj'].fillna(0)
```

```python
matches['diff_surface_type_win_pct'] = matches['higher_ranked_wpct_on_surface_adj'] - matches['lower_ranked_wpct_on_surface_adj']
```

```python
# higher ranked rank, diff_games_dif_per_match, diff_h2h (fill nas), diff_matches_per_day_rate,  diff_recent_win_pct, diff_recent_games_diff, diff_adjusted_surface, diff_rank 
```

```python
# fill head to head nas
matches['diff_h2h_games_dif_per_match'] = matches['diff_h2h_games_dif_per_match'].fillna(matches['diff_h2h_games_dif_per_match'].mean())
```

```python
# fill head to head nas
matches['higher_ranked_h2h_games_dif_per_match'] = matches['higher_ranked_h2h_games_dif_per_match'].fillna(matches['higher_ranked_h2h_games_dif_per_match'].mean())
```

```python
matches['diff_h2h_games_dif_per_match'] = matches['diff_h2h_games_dif_per_match'].fillna(matches['diff_h2h_games_dif_per_match'].mean())
```

```python
matches['diff_rank_performance_game_dif'] = matches['diff_rank_performance_game_dif'].fillna(matches['diff_rank_performance_game_dif'].mean())
```

```python
matches['higher_ranked_rank_performance_game_dif'] = matches['higher_ranked_rank_performance_game_dif'].fillna(matches['higher_ranked_rank_performance_game_dif'].mean())
```

```python
matches['lower_ranked_rank_performance_game_dif'] = matches['lower_ranked_rank_performance_game_dif'].fillna(matches['lower_ranked_rank_performance_game_dif'].mean())
```

```python
matches[['diff_rank', 'diff_recent_games_dif_per_match', 'diff_h2h_games_dif_per_match', 'diff_matches_per_day_rate', 'diff_surface_type_win_pct', 'diff_rank_performance_game_dif', 'lower_ranked_games_dif_per_match', 'higher_ranked_games_dif_per_match', 'lower_ranked_num_champs']].corr()
```

```python
from sklearn.model_selection import train_test_split

X = matches[['diff_rank', 'diff_recent_games_dif_per_match', 'diff_h2h_games_dif_per_match', 'diff_matches_per_day_rate', 'diff_surface_type_win_pct', 'diff_rank_performance_game_dif', 'diff_games_dif_per_match', 'diff_num_champs']]
#X = matches[['lower_ranked_rank', 'higher_ranked_rank', 'higher_ranked_recent_games_dif_per_match','lower_ranked_recent_games_dif_per_match', 'higher_ranked_h2h_games_dif_per_match', 'higher_ranked_matches_per_day_rate', 'lower_ranked_matches_per_day_rate', 'higher_ranked_wpct_on_surface_adj','lower_ranked_wpct_on_surface_adj', 'higher_ranked_rank_performance_game_dif', 'lower_ranked_rank_performance_game_dif', 'lower_ranked_games_dif_per_match', 'higher_ranked_games_dif_per_match', 'lower_ranked_num_champs', 'higher_ranked_num_champs']]

y = matches['HR_win?']


test_size = 0.2
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=123)
```

```python
import statsmodels.api as sm
lr = sm.Logit(y_train, X_train).fit()
```

```python
lr.summary()
```

```python
probs = lr.predict(X_test)
```

```python
prob_df = pd.DataFrame(index=X_test.index, data=probs)
```

```python
prob_df
```

```python
winners = prob_df.merge(matches.loc[X_test.index, ['HR_win?']], left_index=True, right_index=True).rename(columns={0: 'Predicted_HR_Win_Prob'})
winners
```

```python
winners['predict_HR_win'] = winners['Predicted_HR_Win_Prob'] > .5
winners['correct_pred'] = winners['predict_HR_win'] == winners['HR_win?']
winners
```

```python
sum(winners['correct_pred'])/len(winners)
```

```python
sum(winners['HR_win?'])/len(winners)
```

```python
from sklearn import metrics
metrics.confusion_matrix(winners['predict_HR_win'], winners['HR_win?'])
```

```python
y_actu = pd.Series(winners['HR_win?'], name='Actual')
y_pred = pd.Series(winners['predict_HR_win'], name='Predicted')
df_confusion = pd.crosstab(y_actu, y_pred, margins=True)
df_confusion
```

```python
pred_df = prob_df.merge(matches.loc[X_test.index, ['lower_ranked_prob', 'lower_ranked_odds', 'higher_ranked_prob', 'higher_ranked_odds', 'HR_win?']], left_index=True, right_index=True).rename(columns={0: 'Predicted_HR_Win_Prob'})
```

```python
pred_df.head()
```

```python
pred_df['higher_diff'] = pred_df['Predicted_HR_Win_Prob'] - pred_df['higher_ranked_prob']
pred_df.head()
```

```python
pred_df['bet'] = (pred_df['higher_diff'] > .05) | (pred_df['higher_diff'] < -.05)
pred_df.sample(5)
```

```python
def score(row):
    
    if row["Predicted_HR_Win_Prob"] > row['higher_ranked_prob'] and row['HR_win?'] == 1:
        return 1
    if row["Predicted_HR_Win_Prob"] < row['higher_ranked_prob'] and row['HR_win?'] == 0:
        return 1
    return 0

pred_df['correct'] = pred_df.apply(score, axis=1)
```

```python
pred_df
```

```python
pred_df["HR_predict?"] = pred_df['higher_diff'] > 0
pred_df
```

```python
print("Beat the Books Pct: {:.2f}".format(pred_df['correct'].sum()/len(pred_df)))
```

```python
pred_df.loc[pred_df['HR_predict?'] == True, 'ten_dollar_earnings'] = (pred_df['correct']*10*pred_df['higher_ranked_odds']) - 10
```

```python
pred_df.loc[pred_df['HR_predict?'] == False, 'ten_dollar_earnings'] = (pred_df['correct']*10*pred_df['lower_ranked_odds']) - 10
```
