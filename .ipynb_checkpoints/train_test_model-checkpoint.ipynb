```python
import pandas as pd
import numpy as np

pd.set_option('max.columns', 100)
```

```python
matches = pd.read_csv('complete_data.csv').drop('Unnamed: 0', axis=1)
matches.head()
```

```python
matches.index = range(len(matches))
matches
```

```python
matches['higher_ranked_wpct_on_surface_adj'] = matches['higher_ranked_surface_type_win_pct'] - matches['higher_ranked_win_pct']
matches['lower_ranked_wpct_on_surface_adj'] = matches['lower_ranked_surface_type_win_pct'] - matches['lower_ranked_win_pct']
```

```python
matches['diff_surface_type_win_pct'] = matches['higher_ranked_wpct_on_surface_adj'] - matches['lower_ranked_wpct_on_surface_adj']
```

```python
from sklearn.model_selection import train_test_split

X = matches[['diff_rank', 'diff_matches_per_day_rate', 'diff_win_pct', 'diff_surface_type_win_pct', 'diff_series_type_win_pct', 'higher_ranked_rank']]
y = matches['HR_win?']

test_size = 0.2
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=42)
```

```python
import statsmodels.api as sm
lr = sm.Logit(y_train, X_train).fit()
```

```python
lr.summary()
```

```python
probs = lr.predict(X_test)
```

```python
prob_df = pd.DataFrame(index=X_test.index, data=probs)
```

```python
pred_df = prob_df.merge(matches.loc[X_test.index, ['lower_ranked_prob', 'higher_ranked_prob', 'HR_win?']], left_index=True, right_index=True)
```

```python
pred_df.head()
```

```python
pred_df['higher_diff'] = pred_df[0] - pred_df['higher_ranked_prob']
pred_df.head()
```

```python
pred_df['bet'] = (pred_df['higher_diff'] > .05) | (pred_df['higher_diff'] < -.05)
pred_df.sample(5)
```

```python
def score(row):
    
    if row[0] > row['higher_ranked_prob'] and row['HR_win?'] == 1:
        return 1
    if row[0] < row['higher_ranked_prob'] and row['HR_win?'] == 0:
        return 1
    return 0

pred_df['correct'] = pred_df.apply(score, axis=1)
```

```python
pred_df
```

```python
print("Beat the Books Pct: {:.2f}".format(pred_df['correct'].sum()/len(pred_df)))
```

```python
bet_df = pred_df[pred_df['bet'] == True]
bet_df.head()
```

```python
print("Beat the Books Pct on Bets with 5% confidence: {:.2f}".format(bet_df['correct'].sum()/len(bet_df)))
```
